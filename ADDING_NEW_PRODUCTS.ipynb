```python
import json
fd = open("record.json",'r')
r = str(fd.read())
fd.close()

record = json.loads(r)
```

```python
r
```

```python
#Adding new item to Inventory
prod_id = str(input("Enter product id:"))
name = str(input("Enter name:"))
price = int(input("Enter Price:"))
quan = int(input("Enter Quantity:"))
brand =(input("Enter  Brand of Product:"))
size = (input("Enter Size:"))
record[prod_id] = {'name': name, 'quantity': quan, 'price': price, 'Brand': brand, 'Size': size}

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```

```python
#Updating item to Inventory
prod_id = str(input("Enter product id:"))
name = record[prod_id]['name']
print("Your Product Name is: ",name)
price = int(input("Enter Updated price:"))
quan = int(input("Enter Updated quantity:"))
brand = (input("Enter Updated Brand:"))
size = (input("Enter Updated Size:"))
record[prod_id] = {'name': name, 'quantity': quan, 'price': price, 'Brand': brand, 'Size': size}

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```

```python
#Deleting the records
prod_id = str(input("Enter product id to be Deleted:"))
print("You are Deleting: ",record[prod_id]['name'])
del record[prod_id] 

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
